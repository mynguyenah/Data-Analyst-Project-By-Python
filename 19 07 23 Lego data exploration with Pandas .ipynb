### Question 1 : What percentage of all licensed sets ever released ware Star Wars themed ?

```python
# First, merge table
merged = lego_sets.merge(parent_themes, left_on='parent_theme', right_on='name')
merged.drop(columns = 'name_y', inplace=True)

merged.head(5)
```

```python
licensed = merged[merged['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])

#star_wars = licensed[licensed['parent_theme']=='Star Wars']
import re
star_wars = licensed.loc[licensed['parent_theme'].str.contains('star wars', flags=re.I, regex=True)]


the_force = star_wars.shape[0]/licensed.shape[0]
the_force
```

### Question 2 : In which year was Star Wars not the most popular licensed theme ( in terms of number of sets released that year) ?

```python
# sort data by year
licensed_sorted = licensed.sort_values('year')

# add new column with value = 1
licensed_sorted['count'] = 1
licensed_sorted.head(5)

summed_df = licensed_sorted.groupby(['year', 'parent_theme']).sum().reset_index

max_df=summed_df.sort_values('count', ascending=False).drop+duplicates(['year'])

max_df.sort_values('year', inplace=True)
```
